```python
import pandas as pd

all_books = pd.read_json('./series_data/all_books.jl', lines=True)
all_books[['titleComplete','series','ratingsCount','reviewsCount','ratingHistogram']]
```

```python
#df['seriesInfo'] = 
all_books['title'] = all_books['title'].astype(str)
all_books['titleComplete'] = all_books['titleComplete'].astype(str)
all_books['positionInSeries'] = all_books.apply(lambda row: row['titleComplete'].replace(row['title'], '').replace('(','').replace(')','').split('#')[-1], axis=1)
def can_cast_to_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def can_cast_to_int(value):
    try:
        int(value)
        return True
    except ValueError:
        return False

# Apply the function to the 'positionInSeries' column to create the 'isInt' column
all_books['isFloat'] = all_books['positionInSeries'].apply(can_cast_to_float)
all_books['titleComplete'][~(all_books['isFloat'])]

all_books.loc[~(all_books['isFloat']),'positionInSeries']= all_books.loc[~(all_books['isFloat']),:].apply(lambda row: row['positionInSeries'].replace(row['title'], '').replace('(','').replace(')','').split('#')[-1], axis=1)
all_books['isFloat'] = all_books['positionInSeries'].apply(can_cast_to_float)

all_books.groupby('isFloat')['ratingsCount'].sum()
#Some I can't figure out series spot idk really care drop them
books = all_books[all_books['isFloat']]

```

```python
import math
import numpy as np
def calcAverage(hist):
    total = 0
    if not isinstance(hist,list):
        return np.nan
    for i in range(1,6):
        total += i*hist[i-1]
    if sum(hist) == 0:
        return np.nan
    return total / sum(hist)

def calcEntropy(hist):
    total = 0 
    
    if not isinstance(hist,list):
        return np.nan
    if sum(hist) == 0:
        return np.nan
    summed = sum(hist)
    for i in hist:
        if i != 0:
            total += -(i/summed)*math.log(i/summed)
    return total
```

```python
books['positionInSeries'] = books['positionInSeries'].astype(float)
books['avgRating'] = books.apply(lambda row: calcAverage(row['ratingHistogram']), axis=1)
books['entropy'] = books.apply(lambda row: calcEntropy(row['ratingHistogram']), axis=1)
```

```python
books.groupby('positionInSeries')[['avgRating','entropy']].mean()
```

```python
books['isInt'] = books['positionInSeries'].apply(lambda x: x.is_integer())
```

```python
grouped_df = books[(books['isInt']) & (books['positionInSeries'] <= 10) & (books['positionInSeries'] > 0)].groupby('positionInSeries')[['avgRating','entropy','ratingsCount']].mean()
grouped_df
```

```python
from scipy import stats
books['group'] = np.nan
books.loc[(books['isInt']) & (books['positionInSeries'] <= 10) & (books['positionInSeries'] > 0) & (~pd.isna(books['avgRating'])),'group'] = 1
books.loc[(books['isInt']) & (books['positionInSeries'] == 1) & (~pd.isna(books['avgRating'])),'group'] = 0

group_A = books[books['group'] == 0]['avgRating']
group_B = books[books['group'] == 1]['avgRating']

# Performing the t-test
t_stat, p_value = stats.ttest_ind(group_A, group_B, equal_var=True)

# Checking the p-value to determine the result
if p_value < 0.05:
    print("There is a statistically significant difference between the two groups (p-value:", p_value, ")")
else:
    print("There is no statistically significant difference between the two groups (p-value:", p_value, ")")

```

```python
# Performing the Mann-Whitney U test
u_stat, p_value = stats.mannwhitneyu(group_A, group_B)

# Checking the p-value to determine the result
if p_value < 0.05:
    print("There is a statistically significant difference between the two groups (p-value:", p_value, ")")
else:
    print("There is no statistically significant difference between the two groups (p-value:", p_value, ")")
```
