```python
import pandas as pd

all_books = pd.read_json('book_best_series2.jl', lines=True)
all_books[['url','title','series_url']]
```

```python
all_books['series_url'].isnull().sum()
missing = list(all_books[all_books['series_url'].isnull()]['url'])
missing
```

```python
import requests
from bs4 import BeautifulSoup
import time

missing = list(all_books[all_books['series_url'].isnull()]['url'])
series_url_m = []
for url in missing:

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Replace the selector with the one you're testing
    series_url_selector = 'h3.Text.Text__title3.Text__italic.Text__regular.Text__subdued a'
    series_url = soup.select_one(series_url_selector)
    
    if series_url:
        print('Series URL:', series_url['href'])
        series_url_m.append(series_url['href'])
    else:
        print('Series URL not found')
        series_url_m.append(None)
    time.sleep(1)
series_url_m
```

```python
all_books['series_url'].loc[all_books['series_url'].isnull()] = series_url_m
```

```python
all_books.to_csv('all_books.csv',index = False)
```

```python
import requests
from bs4 import BeautifulSoup

# The URL you're trying to scrape
url = 'https://www.goodreads.com/series/45175-harry-potter'

# Fetch the page
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Correct CSS selector to find all <a> tags within the specified <div>
series_url_selector = 'div.responsiveBook__media a'

# Use the selector to find all matching <a> tags
a_tags = soup.select(series_url_selector)

# Iterate over the found <a> tags and extract the 'href' attributes
for a_tag in a_tags:
    series_url = a_tag['href']
    print("Extracted series URL:", series_url)
response.content.css('div.responsiveBook__media a::attr(href)')
```
